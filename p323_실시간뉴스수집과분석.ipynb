(2) url 요청

```
url <- "https://news.daum.net/"
web <- GET(url)
web
```

(3) HTML 파싱
- 웹문서소스를 HTML문서로 파싱하는 함수 : htmlTreeParse()
- 주요 속성
  * uselnternalNodes : 소스 내부의 노드로 변환 여부 지정
  * trim : 텍스트의 맨 앞과 맨 뒤의 공백 제거 여부 지정
  * encoding : 문자셋 인코딩 방식(utf-8, euc-kr)

```
html <- htmlTreeParse(web, useInternalNodes = T, trim = T, encoding = "utf-8")
str(html)
```

```
rootNode <- xmlRoot(html) # 최상위 노드 찾기
# rootNode
```

(4) 태그(tag) 자료 수집
- 크롬 브라우저를 이용하여 해당 URL에 접속함
- 단축키 F12
- 해당 기사에서 오른 마우스하고 [검사] 클릭
- 해당 영역을 감싸고 있는 태그(a)와 속성(class = 'link_txt')을 기록

```
news <- xpathSApply(rootNode,"//a[@class = 'link_txt']",xmlValue)
news
```

(5) 수집한 자료 전처리
- 특수문자, 불용어 제거
- gsub("패턴","교체문자",자료)

단계1: 자료 전처리하기

```
news_pre <- gsub("[\r\n\t]","",news)  # 이스케이프 제거
news_pre <- gsub("[[:punct:]]","",news_pre)  # 문장부호 제거
news_pre <- gsub("[[:cntrl:]]","",news_pre)  # 특수문자 제거
# news_pre <- gsub("\\d+","",news_pre)  # 숫자 제거 (코로나19 숫자 유지)
news_pre <- gsub("[a-z]+","",news_pre)  # 영문자 제거
news_pre <- gsub("[A-Z]+","",news_pre)  # 영문자 제거
news_pre <- gsub("\\s+","",news_pre)    # 2개 이상 공백 교체
news_pre
length(news_pre)
```

단게2:기사와 관계없는 내용 제거

```
news_data <- news_pre[1:44]
news_data
```
